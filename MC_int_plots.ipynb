```python
num_grid=5

x_grid=np.linspace(0.,1.,num_grid, endpoint=False)
y_grid=np.linspace(0.,1.,num_grid,endpoint=False)

x_grid+=1./(num_grid*2)
y_grid+=1./(num_grid*2)


def func(x,y):
    ret= (x**2 + y**2)/((1.-x+1e-10)*(1.-y+1e-10))
    return ret

max_func=0.

for xi in x_grid:
    for yi in y_grid:
        if func(xi,yi)>max_func:
            max_func=func(xi,yi)

            
plt.figure(dpi=150,figsize=[4.,4.])
ax = plt.gca()
plt.xlim(0., 1.)
plt.ylim(0., 1.)

for xi in x_grid:
    for yi in y_grid:
        x_coord=xi-1./(num_grid*2)
        y_coord=yi-1./(num_grid*2)
        rec = Rectangle(xy=(x_coord, y_coord), width=1./(num_grid), height=1./(num_grid),edgecolor='k', fc=str(1.-(func(xi,yi)/max_func)), lw=0.)
        ax.add_patch(rec)

plt.savefig("pixel_"+str(num_grid)+".pdf",bbox_inches='tight')
plt.show()
```
