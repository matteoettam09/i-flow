```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
f_our = open('3jweights.txt','r')
f_sherpa = open('Sherpa.dat','r')
#f_stefan = open('','r')

lines = f_sherpa.readlines()

w_our = []
w_sherpa0 =[]
w_sherpa1 =[]
for line in f_our:
    w_our.append(np.float(line))
w_our = np.array(w_our)

for line in lines[1:]:
    p = line.split()
    w_sherpa0.append(np.float(p[0]))
    w_sherpa1.append(np.float(p[1]))

w_sherpa = np.array([w_sherpa0,w_sherpa1])
xsec_our = 4893.400928408015

# our 4893.400928408015 +/- 4.633143660544801
# sherpa 4877.01 pb +- ( 17.7023 pb = 0.36 \% )
# 12.9 vs 2.8
```

```python
len_sherpa = np.sum(w_sherpa[1])
len_ours = len(w_our)
print(np.sum(w_sherpa[1]))
print(len(w_our))
```

```python
print(w_our/xsec_our)
```

```python
plt.plot(w_sherpa[0],w_sherpa[1])
```

```python
plt.figure(dpi=150,figsize=[6.,4.])
ax = plt.gca()

plt.hist(w_our, bins=np.logspace(np.log10(np.min(w_our))-1e-2,np.log10(np.max(w_our))+1e-2,100),label='Our code')
plt.yscale('log')
plt.xscale('log')
plt.xlim(1e2,5e5)
ax.axes.yaxis.set_ticklabels([])
ax.axes.xaxis.set_ticklabels([])

plt.plot((10**w_sherpa[0])*3.89e8,w_sherpa[1]*len_ours/len_sherpa,color='r',label='Sherpa')
plt.legend()
plt.savefig("weights.pdf",bbox_inches='tight')
plt.show()
```
